```python
#recursive but Too slow :(
def solution(budgets, M):
    answer = 0
    
    budget_len = len(budgets)
    X = int(M/budget_len) #middle point
    bud_sort = sorted(budgets)
    total = 0
    i = 0
    bud_max = max(budgets)
    bud_total = sum(budgets)
    print(bud_total)
    if bud_total <= M:
        return bud_max
    
    
    
    while i < budget_len:
        if bud_sort[i] <= X:
            total += bud_sort[i]
        else:
            total += X
        i += 1
    
    
    if total <= M:
        answer = first(bud_sort, X+1, M)
    else:
        answer = second(bud_sort, X-1, M)
    
    
    return answer

def first(bud, x, m):
    i = 0
    inner_total = 0
    while i < len(bud):
        if bud[i] <= x:
            inner_total += bud[i]
        else:
            inner_total += x
        i += 1
    if inner_total <= m:
        return first(bud, x+1, m)
    else:
        return x-1
    

def second(bud, x, m):
    i = 0
    inner_total = 0
    while i < len(bud):
        if bud[i] <= x:
            inner_total += bud[i]
        else:
            inner_total += x
        i += 1
    if inner_total > m:
        return first(bud, x-1, m)
    else:
        return x
    
    
```
